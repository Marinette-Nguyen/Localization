```python
import os
import pandas as pd
from openpyxl import load_workbook, Workbook
```

### MERGE ALL DEMAND

```python


def merge_excel_files():
    # Define file paths
    import_path = r"C:\Users\Surface\OneDrive\Desktop\python\Localization\Import"
    output_path = r"C:\Users\Surface\OneDrive\Desktop\python\Localization\Output"
    
    path_all_dm_TC2 = os.path.join(import_path, "Forecast_TC2.xlsx")
    path_all_dm_TN2 = os.path.join(import_path, "Forecast_TN2.xlsx")
    path_all_dm = os.path.join(output_path, "All_demand.xlsx")
    
    # Read the Excel files with header from row 3 (header=2 in zero-index)
    df_all_demand_TC2 = pd.read_excel(path_all_dm_TC2, header=2)
    df_all_demand_TN2 = pd.read_excel(path_all_dm_TN2, header=2)

    print(df_all_demand_TC2)
    # Concatenate the dataframes
    merged_df = pd.concat([df_all_demand_TC2, df_all_demand_TN2], ignore_index=True)
    merged_df = merged_df.dropna(subset=["BOM_TYPE"])
    # Load the workbook for Forecast_TC2.xlsx to get cell A1 and header row
    wb_TC2 = load_workbook(path_all_dm_TC2, data_only=True)
    ws_TC2 = wb_TC2.active
    
    # Get the value of cell A1 (assumes it's already a string or a value you can slice)
    cell_A1_value = ws_TC2['A1'].value
    # If cell_A1_value is not a string, convert it:
    if not isinstance(cell_A1_value, str):
        cell_A1_value = str(cell_A1_value)
    
    # Extract a substring from cell A1's value (example: characters from index 50 onward)
    COT = cell_A1_value[50:]
    print(COT)
    
    # Now load the workbook again (or reuse wb_TC2) to extract the header row
    # Here, we can reuse ws_TC2 if needed, but note that iter_rows has already been used, so re-create iterator.
    header_row = [cell.value for cell in next(ws_TC2.iter_rows(min_row=3, max_row=3))]
    
    # Create a new workbook for output
    new_wb = Workbook()
    new_ws = new_wb.active
    new_ws.append(header_row)
    
    # Append merged data rows from the concatenated dataframe
    for row in merged_df.itertuples(index=False, name=None):
        new_ws.append(row)
    
    # Save the new workbook to the output path
    new_wb.save(path_all_dm)
    print(f"Merged file saved to {path_all_dm}")

# Run the function
merge_excel_files()
```
